# classified data analysis (we classify the data here as  their state wise and thus proceed forward for the analysis)

```python
df1=(df.groupby(["State"]).mean()).T
```

```python
print(df1)
```

```python
df1.plot()
```

```python
c1=[]; c2=[];c3=[]
n1=[]; n2=[];n3=[]
f1=[]; f2=[];f3=[]
ex=[]
for i in np.arange(len(df["Profit"])):
    if (df["State"][i])=="California":
        c1=np.append(c1,df["R&D Spend"][i])
        c2=np.append(c2,df["Marketing Spend"][i])
        c3=np.append(c3,df["Profit"][i])
    elif df["State"][i]=="New York":
        n1=np.append(n1,df["R&D Spend"][i])
        n2=np.append(n2,df["Marketing Spend"][i])
        n3=np.append(n3,df["Profit"][i])
    elif df["State"][i]=='Florida':
        f1=np.append(f1,df["R&D Spend"][i])
        f2=np.append(f2,df["Marketing Spend"][i])
        f3=np.append(f3,df["Profit"][i])
    else:
        ex=np.append(ex,0)
```

```python
df_n=DataFrame([n1,n2,n3],index=["R&D Spend","Marketing Spend","Profit"]).T
```

```python
df_f=DataFrame([f1,f2,f3],index=["R&D Spend","Marketing Spend","Profit"]).T
```

```python
df_c=DataFrame([c1,c2,c3],index=["R&D Spend","Marketing Spend","Profit"]).T
```
